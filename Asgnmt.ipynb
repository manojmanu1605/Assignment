```python
def findLength(nums):
    if not nums:
        return 0
    
    max_length = 1
    current_length = 1
    
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_length += 1
        else:
            max_length = max(max_length, current_length)
            current_length = 1
    
    return max(max_length, current_length)

# Test cases
print(findLength([1,3,5,4,7]))  # Output: 3
print(findLength([2,2,2,2,2]))  # Output: 1
```
